```python
import pandas as pd
import numpy as np

pd.set_option('display.max_colwidth', None)
```

```python
df_circles = pd.read_json('Circles.jsonl', lines=True)
df_balloonerism = pd.read_json('Balloonerism.jsonl', lines=True)
df_swimming = pd.read_json('Swimming.jsonl', lines=True)
```

```python
# Concatenate all three DataFrames into a single one
df = pd.concat([df_swimming, df_circles, df_balloonerism], ignore_index=True)
df
```

```python
df.info()
df.count()
df2 = df.drop(columns=["language"])
df=df2
df.info()
df.count()
df
```

```python
missing_values = ["Lyrics not found.", "Error fetching lyrics.", np.nan, None, ""]
df_cleaned = df[~df['lyrics'].isin(missing_values)].copy()
df_cleaned = df_cleaned[df_cleaned['lyrics'].str.strip() != ''].copy()
df_cleaned

```

```python
print(f"Shape of DataFrame after cleaning: {df_cleaned.shape}")
print("\nPreview of the cleaned data:")
df_cleaned.head()
```

```python
df_cleaned['lyrics']=df_cleaned['lyrics'].str.replace(r'\[.*?\]|\(.*?\)', '', regex=True)
df_cleaned.head()
```

```python
df_cleaned.info()
```

```python
#manually adding album release and recorded date
album_data = [
    {
        "album": "Swimming",
        "release_year": 2018,
        "recorded_year": 2018
    },
    {
        "album": "Circles",
        "release_year": 2020,
        "recorded_year": 2018
    },
    {
        "album": "Balloonerism",
        "release_year": 2025,
        "recorded_year": 2014
    }
]
df_albums = pd.DataFrame(album_data)
df_final = pd.merge(df_cleaned, df_albums, on="album", how="left")
```

```python
df_final['lyrics'].str.lower()
```

```python
#cleaning the lyrics column to remove tags, newlines, etc
df_final['lyrics'] = df_final['lyrics'].str.replace(r'\[.*?\]', '', regex=True)
df_final['lyrics'] = df_final['lyrics'].str.replace(r"[^\w\s']", '', regex=True)
df_final['lyrics'] = df_final['lyrics'].str.replace(r'\s+', ' ', regex=True)
df_final['lyrics'] = df_final['lyrics'].str.strip()
print("Created the full cleaned lyrics column for whole-song analysis.")
```

```python
import pandas as pd
import google.generativeai as genai
import os
import json
import time


from google.colab import userdata
api_key = userdata.get('GEMINI_API_KEY')

# Initialize the Gemini client
genai.configure(api_key=api_key)


def get_top_3_emotions_gemini(lyrics):
    time.sleep(1) #to not overload api with too many requests

    if not isinstance(lyrics, str) or len(lyrics.split()) < 3:
        return "neutral", 0.0, "neutral", 0.0, "neutral", 0.0

    emotion_set = [
        "Joy", "Hopefulness", "Calmness / Serenity", "Nostalgia",
        "Love / Affection", "Melancholy", "Sadness / Grief", "Resignation",
        "Frustration", "Longing", "Anger", "Fear / Anxiety"
    ]

    #JSON schema to force a structured output
    schema = {
        "type": "OBJECT",
        "properties": {
            "emotion_1": {"type": "STRING", "enum": emotion_set},
            "score_1": {"type": "NUMBER"},
            "emotion_2": {"type": "STRING", "enum": emotion_set},
            "score_2": {"type": "NUMBER"},
            "emotion_3": {"type": "STRING", "enum": emotion_set},
            "score_3": {"type": "NUMBER"}
        },
        "required": ["emotion_1", "score_1", "emotion_2", "score_2", "emotion_3", "score_3"]
    }

    safety_settings = {
        'HARM_CATEGORY_HARASSMENT': 'BLOCK_NONE',
        'HARM_CATEGORY_HATE_SPEECH': 'BLOCK_NONE',
        'HARM_CATEGORY_SEXUALLY_EXPLICIT': 'BLOCK_NONE',
        'HARM_CATEGORY_DANGEROUS_CONTENT': 'BLOCK_NONE',
    }


    model = genai.GenerativeModel(
        model_name="gemini-1.5-flash",
        generation_config={
            "response_mime_type": "application/json",
            "response_schema": schema
        },
        safety_settings=safety_settings
    )

    prompt = f"""
    You are analyzing a song in Mac miller's album 'Swimming' for deep emotional content. Do NOT focus on surface-level positive words or bravado. Instead, analyze the underlying psychological and emotional state.
.
Focus only on these emotions:

- Joy → feelings of happiness, playfulness, or light-hearted energy.
- Hopefulness → optimism about the future, anticipation of something better.
- Calmness / Serenity → peacefulness, acceptance, or a sense of inner stillness.
- Nostalgia → longing for the past, bittersweet remembrance.
- Love / Affection → intimacy, tenderness, warmth toward self or others.
- Melancholy → reflective, lingering sorrow or wistfulness, not acute sadness.
- Sadness / Grief → acute emotional pain tied to loss or absence.
- Resignation → acceptance of struggle, limitation, or inevitability.
- Frustration → dissatisfaction, feeling stuck, blocked or hindered, tension.
- Longing → yearning for something unattainable, missing, or distant.
- Anger → irritation, resentment, or hostility.
- Fear / Anxiety → worry, dread, unease, or apprehension about uncertainty.

**Instructions:**
- Use the emotion definitions above to guide your choices.
- ANALYZE THESE LYRICS FOR THE TOP 3 EMOTIONS from this allowed list: {', '.join(emotion_set)}.
- Provide a confidence score for each emotion from 0.0 to 1.0.

Lyrics:
"{lyrics}"

    """

    for attempt in range(4):
        try:
            response = model.generate_content(prompt)
            response_data = json.loads(response.text)
            return (
                response_data.get("emotion_1", "error"),
                response_data.get("score_1", 0.0),
                response_data.get("emotion_2", "error"),
                response_data.get("score_2", 0.0),
                response_data.get("emotion_3", "error"),
                response_data.get("score_3", 0.0)
            )
        except Exception as e:
            if attempt == 3:
                print(f"API call failed after multiple retries for lyrics: '{lyrics[:50]}...'. Error: {e}")
                return "api_error", 0.0, "api_error", 0.0, "api_error", 0.0

            wait_time = 2 ** attempt
            print(f"An error occurred: {e}. Retrying in {wait_time} second(s)...")
            time.sleep(wait_time)


#calling the emotion analysis function for the dataframe
unique_songs_df = df_swimming[['track_title', 'lyrics']].drop_duplicates().reset_index(drop=True)
print(f"Analyzing {len(unique_songs_df)} unique songs with Gemini...")

emotion_results = unique_songs_df['lyrics'].apply(
    lambda x: pd.Series(get_top_3_emotions_gemini(x),
                        index=['song_emotion_1', 'song_score_1', 'song_emotion_2', 'song_score_2', 'song_emotion_3', 'song_score_3'])
)

unique_songs_with_emotions = pd.concat([unique_songs_df, emotion_results], axis=1)

df_swimming = pd.merge(
    df_swimming,
    unique_songs_with_emotions[['track_title', 'song_emotion_1', 'song_score_1', 'song_emotion_2', 'song_score_2', 'song_emotion_3', 'song_score_3']],
    on='track_title',
    how='left'
)


```
